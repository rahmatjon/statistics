### **Коэффициент корреляции Пирсона**

Коэффициент корреляции Пирсона — это мера линейной зависимости между двумя переменными. 

Его значения могут варьироваться от -1 до 1, где:

- \( 1 \) — полная положительная линейная зависимость,
- \( -1 \) — полная отрицательная линейная зависимость,
- \( 0 \) — отсутствие линейной зависимости.

### Формула для расчета коэффициента корреляции Пирсона выглядит так:
$$r = \frac{\sum{(X - \bar{X})(Y - \bar{Y})}}{\sqrt{\sum{(X - \bar{X})^2}} \cdot \sqrt{\sum{(Y - \bar{Y})^2}}}$$

### Пример

Рассмотрим, например, зависимость между двумя переменными: X и Y.

#### Шаги:
1. Рассчитать коэффициент корреляции Пирсона.
2. Построить график для визуализации зависимости между переменными.
3. Нарисуем на графике линию тренда.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression

# Генерация данных
np.random.seed(42)
x = np.random.rand(100) * 10  # 100 случайных точек от 0 до 10
y = 2 * x + np.random.randn(100) * 2  # линейная зависимость с шумом

# Вычисление коэффициента корреляции Пирсона
corr, _ = pearsonr(x, y)
print(f"Коэффициент корреляции Пирсона: {corr:.2f}")

# Построение графика
plt.scatter(x, y, label="Данные", color="blue")

# Линия тренда
x_reshaped = x.reshape(-1, 1)
model = LinearRegression()
model.fit(x_reshaped, y)
y_trend = model.predict(x_reshaped)

plt.plot(x, y_trend, color='red', label="Линия тренда")
plt.title(f"График данных и линия тренда\nКоэффициент корреляции Пирсона: {corr:.2f}")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

### Объяснение кода:
- `pearsonr(X, Y)` вычисляет коэффициент корреляции Пирсона и его p-значение.
- `plt.scatter(X, Y)` строит диаграмму рассеяния для визуализации данных.

### **Коэффициент корреляции Спирмена**

Коэффициент корреляции Спирмена используется для измерения монотонной зависимости между двумя переменными. Он оценивает, насколько одна переменная может быть выражена как монотонная функция другой переменной, и применяется, когда зависимость не обязательно линейна.

Вот пример, как рассчитать коэффициент корреляции Спирмена на Python с использованием библиотеки `scipy`, а также построить график рассеяния для двух переменных.

### Шаги:
1. Рассчитаем коэффициент корреляции Спирмена с помощью функции `spearmanr` из библиотеки `scipy.stats`.
2. Построим график рассеяния с помощью `matplotlib`.

### Пример кода:
Предположим, у нас есть две выборки данных, и мы хотим вычислить коэффициент корреляции Спирмена и построить график с линией тренда.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr

# Пример данных
x = np.array([10, 20, 28, 40, 50])
y = np.array([8, 18, 32, 45, 48])

# Расчет коэффициента корреляции Спирмена
corr, _ = spearmanr(x, y)
print(f"Коэффициент корреляции Спирмена: {corr:.2f}")

# Создание DataFrame для удобства визуализации
df = pd.DataFrame({'x': x, 'y': y})

# Построение графика рассеяния
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', data=df, s=100, color='b', label='Данные')

# Линия тренда с помощью полиномиальной аппроксимации (1-й степени, т.е. линейная)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--", label='Линия тренда')

# Настройки графика
plt.title('График зависимости с линией тренда')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

### Объяснение кода:
1. **Данные**: В примере используются два массива данных `x` и `y`.
2. **Коэффициент Спирмена**: Используется функция `spearmanr` из библиотеки `scipy.stats` для вычисления коэффициента корреляции.
3. **График**: Построен график рассеяния (scatterplot) с линией тренда, которая создается с помощью полиномиальной аппроксимации первого порядка (линейной).
