```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate a sample dataset
# np.random.seed(0) # <= Test it
data = np.random.normal(0, 20, 1000)
data[:100]
```

```python
# Generate a sample dataset
help(np.random.normal)
```

```python
# Create a histogram to get the bin counts and bin edges
counts, bin_edges, _ = plt.hist(data, bins=60, alpha=0.5, color='gray', edgecolor='black', label='Histogram')
# counts, bin_edges, _ = plt.hist(data, alpha=0.5, color='gray', edgecolor='black', label='Histogram')

# Add labels and title
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Calculate the midpoints of the bins
bin_midpoints = (bin_edges[:-1] + bin_edges[1:]) / 2
print("bin_edges =>")
print(bin_edges)
print("bin_edges[:-1] =>")
print(bin_edges[:-1])
print("bin_edges[1:] =>")
print(bin_edges[1:])
print("bin_midpoints =>")
print(bin_midpoints)
```

```python
# Plot the frequency polygon
plt.plot(bin_midpoints, counts, marker='o', linestyle='-', color='b', label='Frequency Polygon')

# Add labels and title
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('Frequency Polygon')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
