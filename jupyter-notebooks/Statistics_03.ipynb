```python
import random
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# coin, heads and tails
# version 1
def coin_trial():
    result = 'heads'
    if random.random() < 0.5:
        result = 'tails'
    return result
```

```python
x = coin_trial()
x
```

```python
def simulate_coins(n):
    trials = []
    for i in range(n):
        trials.append(coin_trial())
    return trials
```

```python
three_coins = simulate_coins(3)
three_coins
```

```python
data = simulate_coins(20000)
```

```python
# Create a histogram to get the bin counts and bin edges
plt.hist(data, bins=2, alpha=0.5, color='gray', edgecolor='black', label='Coin Trials')

# Add labels and title
plt.xlabel('Coin sides')
plt.ylabel('Frequency')
plt.title('Coin Trials Modelling')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
df = pd.DataFrame(data, columns=['coins'])
df
```

```python
sides_counts = df['coins'].value_counts()
sides_counts
```

```python
type(sides_counts)
```

```python
# Create a DataFrame
df1 = pd.DataFrame([sides_counts])
df1
```

```python
# Calculating Percentage
df1['heads_percentage'] = df1['heads'] / (df1['heads'] + df1['tails']) * 100
df1['tails_percentage'] = df1['tails'] / (df1['heads'] + df1['tails']) * 100
df1
```
