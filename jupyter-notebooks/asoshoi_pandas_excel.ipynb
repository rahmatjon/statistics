Ба df сутуни нав (summa = volume * price) илова мекунем:

```python
df['summa'] = round(df['price'] * df['volume'], 2)
df
```

Барои майдони 'summa' минимум, максимум, миёнаи арифметикӣ ва суммаро ҳисоб мекунем:

```python
print("Минимум: ",df["summa"].min())
print("Максимум: ",df["summa"].max())
print("Миёнаи арифметикӣ: ",df["summa"].mean())
print("Сумма: ",df["summa"].sum())
```
