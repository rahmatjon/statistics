## Однофакторный дисперсионный анализ (ANOVA)

Однофакторный дисперсионный анализ (ANOVA) используется для сравнения средних значений трёх или более групп для проверки гипотезы о равенстве их средних. Процедура однофакторного ANOVA включает следующие шаги:

### Процедура однофакторного ANOVA:
1. **Формулировка гипотез:**
   - **Нулевая гипотеза ($H_0$):** Средние значения всех групп равны.
   - **Альтернативная гипотеза ($H_1$):** Хотя бы одно из средних значений отличается.

2. **Расчёт общей дисперсии (межгрупповой и внутригрупповой):**
   - **Межгрупповая дисперсия** отражает различия между средними значениями групп.
   - **Внутригрупповая дисперсия** отражает вариации внутри каждой группы.

3. **Вычисление F-статистики:**
   F = (Межгрупповая дисперсия / Внутригрупповая дисперсия)

4. **Проверка F-статистики по таблице критических значений:**
   Сравнивается полученное значение F с критическим значением для заданного уровня значимости (обычно α = 0.05) и степеней свободы.

5. **Принятие решения:**
   - Если F > критическое значение, отвергаем нулевую гипотезу.
   - Если F ≤ критическое значение, нулевая гипотеза не отвергается.

### Пример кода ANOVA на Python:
Для выполнения однофакторного ANOVA можно использовать библиотеку `scipy` и построить график с помощью `matplotlib`.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Данные для примера
group1 = [20, 22, 19, 23, 21]
group2 = [30, 32, 29, 31, 30]
group3 = [40, 42, 41, 43, 44]

# Однофакторный ANOVA
f_stat, p_value = stats.f_oneway(group1, group2, group3)

print(f'F-статистика: {f_stat}')
print(f'p-значение: {p_value}')

# Построение графика
data = [group1, group2, group3]
plt.boxplot(data, tick_labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Распределение значений в группах')
plt.ylabel('Значения')
plt.show()
```

### Описание кода:
- Используются три группы данных `group1`, `group2`, и `group3`.
- Функция `stats.f_oneway()` из библиотеки `scipy.stats` выполняет однофакторный ANOVA и возвращает F-статистику и p-значение.
- Для визуализации распределений используется диаграмма размаха (boxplot).

### Интерпретация результатов:
- **F-статистика**: Чем выше значение, тем больше различие между группами.
- **p-значение**: Если оно меньше уровня значимости (например, 0.05), можно отвергнуть нулевую гипотезу и сделать вывод о том, что средние значения групп различаются.
