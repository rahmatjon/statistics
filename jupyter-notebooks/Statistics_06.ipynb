<p>We may easily generate <code>n</code> number of random samples within the scopes of a mean (<code>loc</code>) and a standard deviation (<code>scale</code>).</p>

We may approximate the density function for $X \sim U(-2, 0.8)$ by using the uniform.rvs() function and plot the results as a histogram.

```python
rand_unif = stats.uniform.rvs(loc=-2, scale=2.8, size=15000)
plt.figure(figsize=(10, 5))
plt.hist(rand_unif, density=True, color="lightgrey", edgecolor="darkgrey")
plt.xlabel("x")
plt.ylabel("Density")
plt.title('Histogram of rand_unif')
plt.show()
```

<p>Further, we plot both, the density histogram from above as well as the uniform probability distribution for the interval [-2,0.8], by applying the <code>unform.pdf()</code> function.</p>

```python
a = -2
b = 0.8

plt.figure(figsize=(10, 5))
plt.hist(rand_unif, density=True, color="lightgrey", edgecolor="darkgrey")
plt.title("Uniform distribution for the interval [-2,0.8]", fontsize =16)

x = np.arange(-5, 5, 0.001)
plt.plot(x, stats.uniform.pdf(x, loc=a, scale=2.8), color="darkblue")

plt.ylim(-0.01, 0.4)
plt.xlim(-3, 3)

plt.xlabel("x", fontsize =14)
plt.ylabel("Density", fontsize =14)
plt.show()
```

<h3><span>Normal Distribution</span></h3>
<p><strong>Normal Distribution</strong><span style="background-color:rgb(255,255,255);color:rgba(0,0,0,0.84);"> is a probability function used in statistics that tells about how the data values are distributed. It is the most important probability distribution function used in statistics because of its advantages in real case scenarios. For example, the height of the population, shoe size, IQ level, rolling a die, and many more.</span>&nbsp;</p>
<p>The probability density function of normal or Gaussian distribution is given by:</p>
<p><b><strong>Probability Density Function:</strong></b></p>
$$f(x) = \dfrac{1}{\sigma\sqrt{2 \pi }} e^{-\dfrac{(x - \mu)^2}{2 \sigma^2}}$$
<p>Where, x is the variable, mu is the mean, and sigma standard deviation</p>
<p dir="ltr"><span>The </span><b><strong>normal distribution chart</strong></b><span> is characterized by two parameters:&nbsp;</span></p>
<ul><li value="1"><span>The average value, which represents the maximum value of the chart, and the chart is always symmetrical.&nbsp;</span></li><li value="2"><span>And the standard deviation, which determines the amount of change beyond the mean. Smaller standard deviations (compared to the mean) appear steeper, while larger standard deviations (compared to the mean) appear flat.</span></li></ul><p dir="ltr"><b><strong>Plotting the Normal Distribution</strong></b></p>
<ul><li value="1"><span>NumPy arange() is used to create and return a reference to a uniformly distributed ndarray instance.&nbsp;</span></li><li value="2"><span>With the help of </span><b><strong>mean() </strong></b><span>and </span><b><strong>stdev() </strong></b><span>method, we calculated the mean and standard deviation and initialized to </span><b><strong>mean </strong></b><span>and </span><b><strong>sd </strong></b><span>variable.&nbsp;</span></li><li value="3"><span>Inside the plot() method, we used one method </span><b><strong>pdf() </strong></b><span>for displaying the probability density function. This </span><b><strong>pdf()</strong></b><span> method present inside the </span><b><strong>scipy.stats.norm.&nbsp;</strong></b></li></ul><p dir="ltr"><b><strong>Example:</strong></b></p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
  
# Plot between -30 and 30 with
# 0.1 steps.
x_axis = np.arange(-30, 30, 0.1)
  
# Calculating mean and standard 
# deviation
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
  
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```

<h3><span>Normal Distribution over Histogram</span></h3><p dir="ltr"><span>Now, we are done separated the histogram and the normal distribution plot discussion, but it would be great if we can visualize them in a graph with the same scale. This can be easily achieved by accessing two charts in the same cell and then using</span><b><strong> plt.show(). </strong></b><span>Now, Let’s discuss about Plotting Normal Distribution over Histogram using Python.&nbsp;</span></p>
<p dir="ltr"><span>We believe that the histogram of some data follows a normal distribution. SciPy has a variety of methods that can be used to estimate the best distribution of </span><a href="https://www.geeksforgeeks.org/random-variable/" rel="noopener" target="_blank"><span>random variables</span></a><span>, as well as parameters that can best simulate this adaptability. For example, for the data in this problem, the mean and standard deviation of the best-fitting normal distribution can be found as follows:</span></p>
<pre><span># Make the normal distribution fit the data: </span><br><span>mu, std = norm.fit (data) # mean and standard deviation</span><br></pre><p dir="ltr"><span>The function xlim() within the Pyplot module of the Matplotlib library is used to obtain or set the x limit of this axis.</span></p>
<blockquote><p dir="ltr"><b><strong>Syntax: </strong></b><span>matplotlib.pyplot.xlim (*args, **kwargs) &nbsp;</span></p>
<p dir="ltr"><b><strong>Parameters: &nbsp;</strong></b><span>This method uses the following parameters, as described below: &nbsp;</span></p>

<ul><li value="1"><b><strong>left, right:</strong></b><span> return a tuple of the new limit value of the x-axis.</span></li></ul></blockquote>

Рост людей в популяции часто подчиняется нормальному распределению. Если средний рост составляет 170 см, а стандартное отклонение — 10 см, можно использовать нормальное распределение для оценки вероятности того, что рост случайно выбранного человека будет в пределах определенного интервала.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate some data for this 
# demonstration.
data = np.random.normal(170, 10, 12500)

# Fit a normal distribution to
# the data:
# mean and standard deviation
mu, std = norm.fit(data) 

# Plot the histogram.
plt.hist(data, bins=25, density=True, alpha=0.6, color='b')

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

plt.plot(x, p, 'k', linewidth=2)
title = "Fit Values: {:.2f} and {:.2f}".format(mu, std)
plt.title(title)

plt.show()
```

```python
exp_dist = stats.expon(scale = 1/0.2)
x = np.arange(0, 10, 0.1)
y = exp_dist.cdf(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Cumulative Distribution Function of Exponential Distribution')
plt.show()
```

```python
exp_dist = stats.expon(scale = 1/0.2)
p = np.arange(0, 1, 0.01)
x = exp_dist.ppf(p)
plt.plot(p, x)
plt.xlabel('p')
plt.ylabel('x')
plt.title('Percent Point Function of Exponential Distribution')
plt.show()
```
