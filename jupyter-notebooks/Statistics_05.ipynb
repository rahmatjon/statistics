### Binomial Distribution

```python
def binomial(x, n, p):
    return math.comb(n, x) * (p ** x) * ((1 - p) ** (n - x))
```

```python
def normal_distribution(x, mu, sigma):
    return (math.exp(-0.5 * ((x - mu) / sigma) ** 2)) / (sigma * math.sqrt(2 * math.pi))
```

```python
n = 10
p = 0.5
binomial_list = []
keys = []
```

```python
for x in range(10):
    binomial_list.append(binomial(x, n, p))

for y in range(10):
    keys.append(y)
```

```python
plt.bar(x=keys, height=binomial_list)
plt.plot(keys, [normal_distribution(y, n*p, math.sqrt(n*p*(1-p))) for y in range(10)])
plt.show()
```

<p>Here is the Python code calculating geometric probability distribution.</p>
<p>Pay attention to some of the following:</p>
<ul class="wp-block-list"><li>Discrete random variable X is defined along with probability of the perfect throw (event to occur)</li><li><a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.geom.html" target="_blank" rel="noreferrer noopener">Scipy.stats geom</a> class is used to calculate the<strong> probability mass function</strong> using the method, <strong>pmf</strong>.</li></ul>



```python
# X = Discrete random variable representing number of throws
# p = Probability of the perfect throw
#
n = 20
X = [_ for _ in range(1,21)]
p = 0.2
#
# Calculate geometric probability distribution
#
geom_pd = geom.pmf(X, p)
#
# Plot the probability distribution
#
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.plot(X, geom_pd, 'bo', ms=8, label='geom pmf')
plt.ylabel("Probability", fontsize="18")
plt.xlabel("X - No. of Throws", fontsize="18")
plt.title("Geometric Distribution - No. of Throws Vs Probability", fontsize="18")
ax.vlines(X, 0, geom_pd, colors='b', lw=5, alpha=0.5)
```

### Hypergeometric Distribution

The symbols used to denote the shape parameters (M, n, and N) are not universally accepted. See the Examples for a clarification of the definitions used here.

The probability mass function is defined as,

$$P(X = k) = \dfrac{\binom{K}{k} \binom{N - K}{n - k}}{\binom{N}{n}}$$

```python
from scipy.stats import hypergeom
```

Suppose we have a collection of 20 animals, of which 7 are dogs. 
Then if we want to know the probability of finding a given number of dogs if we choose at random 12 of the 20 animals, we can initialize a frozen distribution and plot the probability mass function:

```python
[M, n, N] = [20, 7, 12]
rv = hypergeom(M, n, N)
x = np.arange(0, n+1)
pmf_array = rv.pmf(x)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_array, 'bo')
ax.vlines(x, 0, pmf_array, lw=2)
ax.set_xlabel('# of dogs in our group of chosen animals')
ax.set_ylabel('hypergeom PMF')
plt.show()
```

В коробке 20 шаров, из которых 7 красные и 13 синие. Если мы выбираем 5 шаров без замены, вероятность того, что из выбранных 5 шаров ровно 3 будут красными:

```python
[M, n, N] = [20, 7, 5]
rv = hypergeom(M, n, N)
x = np.arange(0, n+1)
pmf_array = rv.pmf(x)
pmf_array
```
